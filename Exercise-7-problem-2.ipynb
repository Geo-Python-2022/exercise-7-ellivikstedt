### Part 2 (2 points)

Select data for the 30-year period (January 1988 to December 2018).

- Store the selection in a new variable `selection`

```python
selection = data.loc[(data.index >= "19880101") & (data.index <= "20181231")]
```

Save your figure as PNG file called `temp_line_plot.png`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Creating base for the figure
fig, axis = plt.subplots(nrows=1, ncols=1, figsize=(14,6))

# Setting axis
ax11 = axis

# Selecting only temperature values in Celsius from the data
temp_c = selection["TEMP_C"]

# Plotting the data and setting parameters 
temp_c.plot(
    ax=ax11, 
    marker=".",
    c="k", lw=1, 
    ylim=[-15, 25],
    title="Helsinki-Vantaa airport",
    xlabel="Time",
    ylabel="Temperature °C",
    grid=True,
)

# Setting output file name
outputfp = "temp_line.plot.png"

# Saving plot as png
plt.savefig(outputfp)
```
